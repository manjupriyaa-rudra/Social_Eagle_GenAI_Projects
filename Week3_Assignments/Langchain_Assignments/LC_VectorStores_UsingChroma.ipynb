```python
# importing the components / methods

from langchain_community.embeddings import HuggingFaceEmbeddings

from langchain_community.vectorstores import Chroma
```

```python
# downloaing the model for embedding

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
# Creating document

texts = [
    "LangChain is a framework for LLM apps.",
    "Chroma is a vector database.",
    "Embeddings convert text into numbers.",
    "India is a beautiful country."
]
```

```python
# Create chroma vector store

db = Chroma.from_texts(texts,embeddings,persist_directory="chroma_db")
```

```python
# Similarity search

results = db.similarity_search("What is chroma?", k=2)

for r in results:
    print(r.page_content)
```

```python
# Storing your vector

db.persist()
```
